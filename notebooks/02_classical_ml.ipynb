```python
import os
import random
import numpy as np
from collections import Counter
import cv2
from PIL import Image
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV, cross_val_score, train_test_split
from skimage.feature import local_binary_pattern, graycomatrix, graycoprops
from skimage.feature import hog
from skimage import color, measure
import matplotlib.pyplot as plt
from tqdm import tqdm
import joblib
import time
import sys

# -----------------
# CONFIG
# -----------------
SEED = 42
IMG_SIZE = 224
random.seed(SEED)
np.random.seed(SEED)

# -----------------
# FEATURE EXTRACTION
# -----------------
class FeatureExtractor:
    def __init__(self, img_size=IMG_SIZE):
        self.img_size = img_size
        
    def resize_and_pad(self, img_path):
        """Same preprocessing as deep learning approach"""
        img = cv2.imread(img_path)
        if img is None:
            return None
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        h, w = img.shape[:2]
        scale = self.img_size / max(w, h)
        new_w, new_h = int(w * scale), int(h * scale)
        
        img = cv2.resize(img, (new_w, new_h))
        
        # Padding
        pad_w, pad_h = self.img_size - new_w, self.img_size - new_h
        top, bottom = pad_h // 2, pad_h - pad_h // 2
        left, right = pad_w // 2, pad_w - pad_w // 2
        img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT, value=0)
        
        return img
    
    def extract_color_features(self, img):
        """Extract color histogram features"""
        features = []
        
        # RGB histograms
        for channel in range(3):
            hist = cv2.calcHist([img], [channel], None, [32], [0, 256])
            features.extend(hist.flatten())
        
        # HSV histograms
        hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        for channel in range(3):
            hist = cv2.calcHist([hsv], [channel], None, [32], [0, 256])
            features.extend(hist.flatten())
        
        # Color moments (mean, std, skewness)
        for channel in range(3):
            channel_data = img[:, :, channel].flatten()
            features.extend([
                np.mean(channel_data),
                np.std(channel_data),
                measure.moments_central(channel_data, order=3)[0] if len(channel_data) > 0 else 0
            ])
            
        return np.array(features)
    
    def extract_texture_features(self, img):
        """Extract texture features using LBP and GLCM"""
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        features = []
        
        # Local Binary Pattern (LBP)
        radius = 3
        n_points = 8 * radius
        lbp = local_binary_pattern(gray, n_points, radius, method='uniform')
        lbp_hist, _ = np.histogram(lbp.ravel(), bins=n_points + 2, range=(0, n_points + 2))
        features.extend(lbp_hist)
        
        # Gray Level Co-occurrence Matrix (GLCM) - reduced for efficiency
        distances = [5]
        angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]
        
        # Downsample for GLCM computation to reduce computational cost
        gray_small = cv2.resize(gray, (64, 64))
        glcm = graycomatrix(gray_small, distances, angles, 256, symmetric=True, normed=True)
        
        # Extract GLCM properties
        for prop in ['dissimilarity', 'correlation', 'homogeneity', 'contrast', 'energy']:
            prop_values = graycoprops(glcm, prop)
            features.extend(prop_values.flatten())
        
        return np.array(features)
    
    def extract_shape_features(self, img):
        """Extract basic shape and geometric features"""
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        
        # Find contours
        _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        features = []
        if contours:
            # Largest contour
            largest_contour = max(contours, key=cv2.contourArea)
            
            # Basic shape features
            area = cv2.contourArea(largest_contour)
            perimeter = cv2.arcLength(largest_contour, True)
            
            # Bounding rectangle
            x, y, w, h = cv2.boundingRect(largest_contour)
            aspect_ratio = float(w) / h if h > 0 else 0
            extent = float(area) / (w * h) if w * h > 0 else 0
            
            # Solidity
            hull = cv2.convexHull(largest_contour)
            hull_area = cv2.contourArea(hull)
            solidity = float(area) / hull_area if hull_area > 0 else 0
            
            features = [area, perimeter, aspect_ratio, extent, solidity]
        else:
            features = [0, 0, 0, 0, 0]
        
        return np.array(features)
    
    def extract_hog_features(self, img):
        """Extract HOG features"""
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        hog_features = hog(gray, orientations=9, pixels_per_cell=(8, 8),
                          cells_per_block=(2, 2), visualize=False)
        return hog_features
    
    def extract_all_features(self, img_path):
        """Extract all features from an image"""
        img = self.resize_and_pad(img_path)
        if img is None:
            return None
        
        try:
            color_features = self.extract_color_features(img)
            texture_features = self.extract_texture_features(img)
            shape_features = self.extract_shape_features(img)
            hog_features = self.extract_hog_features(img)
            
            # Combine all features
            all_features = np.concatenate([color_features, texture_features, 
                                         shape_features, hog_features])
            return all_features
        except Exception as e:
            print(f"Error extracting features from {img_path}: {e}")
            return None

# -----------------
# HIERARCHICAL CLASSICAL ML CLASSIFIERS
# -----------------
class HierarchicalClassicalML:
    def __init__(self, data_dir, class_names):
        self.data_dir = data_dir
        self.class_names = class_names
        self.feature_extractor = FeatureExtractor()
        
        # Map classes according to the hierarchical structure
        self.healthy_idx = class_names.index("Healthy")  # Index 2
        self.weed_idx = class_names.index("Shepherd_purse_weeds")  # Index 5
        self.disease_indices = [0, 1, 3, 4, 6, 7]  # All disease classes
        
        # Models
        self.classifier1 = None  # Crop vs Weed
        self.classifier2 = None  # Healthy vs Disease
        self.classifier3 = None  # Disease types
        
        # Scalers
        self.scaler1 = StandardScaler()
        self.scaler2 = StandardScaler()
        self.scaler3 = StandardScaler()
        
        print(f"🎯 HIERARCHICAL STRUCTURE INITIALIZED:")
        print(f"   Healthy class: {class_names[self.healthy_idx]} (index: {self.healthy_idx})")
        print(f"   Weed class: {class_names[self.weed_idx]} (index: {self.weed_idx})")
        print(f"   Disease classes: {[class_names[i] for i in self.disease_indices]} (indices: {self.disease_indices})")

    def load_and_extract_features(self):
        """Load images and extract features"""
        print("\n" + "="*70)
        print("FEATURE EXTRACTION PHASE")
        print("="*70)
        
        all_features = []
        all_labels = []
        class_counts = {}
        failed_extractions = 0
        
        total_images = 0
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(self.data_dir, class_name)
            if os.path.exists(class_dir):
                class_image_count = len([f for f in os.listdir(class_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
                total_images += class_image_count
        
        print(f"🖼️  Found {total_images} total images across {len(self.class_names)} classes")
        print(f"📏 Target image size: {self.feature_extractor.img_size}x{self.feature_extractor.img_size}")
        print("\nProcessing classes:")
        
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(self.data_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"⚠️  Directory not found: {class_dir}")
                continue
            
            class_features = []
            class_failed = 0
            
            print(f"\n📁 Processing {class_name}...")
            image_files = [f for f in os.listdir(class_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
            
            for img_file in tqdm(image_files, desc=f"  Extracting features"):
                img_path = os.path.join(class_dir, img_file)
                features = self.feature_extractor.extract_all_features(img_path)
                
                if features is not None:
                    all_features.append(features)
                    all_labels.append(class_idx)
                    class_features.append(features)
                else:
                    class_failed += 1
                    failed_extractions += 1
            
            # Class statistics
            successful_extractions = len(class_features)
            class_counts[class_name] = successful_extractions
            
            if successful_extractions > 0:
                class_features = np.array(class_features)
                feature_stats = {
                    'mean': np.mean(class_features, axis=0),
                    'std': np.std(class_features, axis=0),
                    'min': np.min(class_features, axis=0),
                    'max': np.max(class_features, axis=0)
                }
                
                print(f"  ✅ Successfully processed: {successful_extractions} images")
                print(f"  📊 Feature vector size: {len(class_features[0])}")
                print(f"  📈 Feature statistics - Mean: {np.mean(feature_stats['mean']):.3f}, Std: {np.mean(feature_stats['std']):.3f}")
                if class_failed > 0:
                    print(f"  ❌ Failed extractions: {class_failed}")
            else:
                print(f"  ❌ No successful feature extractions for {class_name}")
        
        # Overall statistics
        print(f"\n" + "="*50)
        print("FEATURE EXTRACTION SUMMARY")
        print("="*50)
        
        if all_features:
            all_features = np.array(all_features)
            all_labels = np.array(all_labels)
            
            print(f"✅ Total successful extractions: {len(all_features)}")
            print(f"❌ Total failed extractions: {failed_extractions}")
            print(f"📊 Final feature matrix shape: {all_features.shape}")
            print(f"🎯 Success rate: {len(all_features)/(len(all_features)+failed_extractions)*100:.1f}%")
            
            # Feature statistics
            print(f"\n📈 FEATURE STATISTICS:")
            print(f"  Mean feature value: {np.mean(all_features):.6f}")
            print(f"  Std feature value: {np.std(all_features):.6f}")
            print(f"  Min feature value: {np.min(all_features):.6f}")
            print(f"  Max feature value: {np.max(all_features):.6f}")
            print(f"  Features with zero variance: {np.sum(np.var(all_features, axis=0) == 0)}")
            
            # Class distribution
            print(f"\n🏷️  CLASS DISTRIBUTION:")
            label_counts = Counter(all_labels)
            for class_idx in sorted(label_counts.keys()):
                class_name = self.class_names[class_idx]
                count = label_counts[class_idx]
                percentage = (count / len(all_labels)) * 100
                print(f"  {class_name:25} | {count:4d} samples ({percentage:5.1f}%)")
            
            # Check for class imbalance
            max_count = max(label_counts.values())
            min_count = min(label_counts.values())
            imbalance_ratio = max_count / min_count if min_count > 0 else float('inf')
            print(f"\n⚖️  Class imbalance ratio: {imbalance_ratio:.1f}:1")
            if imbalance_ratio > 10:
                print("  ⚠️  Severe class imbalance detected - hierarchical approach recommended!")
            
        else:
            print("❌ No features extracted successfully!")
            return None, None
        
        print("="*50)
        return all_features, all_labels

    def prepare_datasets(self, features, labels, test_size=0.2):
        """Prepare datasets for hierarchical classification"""
        print(f"\n🔄 PREPARING HIERARCHICAL DATASETS (Test size: {test_size*100:.0f}%)")
        print("-" * 50)
        
        # Train-test split
        X_train, X_test, y_train, y_test = train_test_split(
            features, labels, test_size=test_size, random_state=SEED, stratify=labels)
        
        print(f"📊 MAIN SPLIT:")
        print(f"  Training samples: {len(X_train)}")
        print(f"  Test samples: {len(X_test)}")
        
        # Dataset 1: Crop vs Weed (binary)
        print(f"\n🌱 DATASET 1: CROP vs WEED")
        y_train_1 = np.where(y_train == self.weed_idx, 0, 1)  # 0: weed, 1: crop
        y_test_1 = np.where(y_test == self.weed_idx, 0, 1)
        
        train_weed_count = np.sum(y_train_1 == 0)
        train_crop_count = np.sum(y_train_1 == 1)
        test_weed_count = np.sum(y_test_1 == 0)
        test_crop_count = np.sum(y_test_1 == 1)
        
        print(f"  Training: Weed={train_weed_count}, Crop={train_crop_count}")
        print(f"  Test: Weed={test_weed_count}, Crop={test_crop_count}")
        
        # Dataset 2: Healthy vs Disease (binary, only crops)
        print(f"\n🏥 DATASET 2: HEALTHY vs DISEASE (crops only)")
        crop_mask_train = y_train != self.weed_idx
        crop_mask_test = y_test != self.weed_idx
        
        X_train_2 = X_train[crop_mask_train]
        X_test_2 = X_test[crop_mask_test]
        y_train_2 = np.where(y_train[crop_mask_train] == self.healthy_idx, 1, 0)  # 1: healthy, 0: disease
        y_test_2 = np.where(y_test[crop_mask_test] == self.healthy_idx, 1, 0)
        
        train_disease_count = np.sum(y_train_2 == 0)
        train_healthy_count = np.sum(y_train_2 == 1)
        test_disease_count = np.sum(y_test_2 == 0)
        test_healthy_count = np.sum(y_test_2 == 1)
        
        print(f"  Training: Disease={train_disease_count}, Healthy={train_healthy_count}")
        print(f"  Test: Disease={test_disease_count}, Healthy={test_healthy_count}")
        
        # Dataset 3: Disease classification (multiclass, only diseases)
        print(f"\n🦠 DATASET 3: DISEASE CLASSIFICATION (disease types only)")
        disease_mask_train = np.isin(y_train, self.disease_indices)
        disease_mask_test = np.isin(y_test, self.disease_indices)
        
        X_train_3 = X_train[disease_mask_train]
        X_test_3 = X_test[disease_mask_test]
        y_train_3 = y_train[disease_mask_train]
        y_test_3 = y_test[disease_mask_test]
        
        # Map disease labels to 0-based indices for classifier 3
        disease_label_map = {disease_idx: i for i, disease_idx in enumerate(self.disease_indices)}
        y_train_3_mapped = np.array([disease_label_map[label] for label in y_train_3])
        y_test_3_mapped = np.array([disease_label_map[label] for label in y_test_3])
        
        print(f"  Training samples: {len(X_train_3)}")
        print(f"  Test samples: {len(X_test_3)}")
        print(f"  Disease classes: {len(self.disease_indices)}")
        
        if len(X_train_3) > 0:
            print("  Disease class distribution in training:")
            disease_train_counts = Counter(y_train_3_mapped)
            for mapped_idx, count in sorted(disease_train_counts.items()):
                original_idx = self.disease_indices[mapped_idx]
                disease_name = self.class_names[original_idx]
                print(f"    {disease_name}: {count} samples")
        
        datasets = {
            'classifier1': (X_train, X_test, y_train_1, y_test_1),
            'classifier2': (X_train_2, X_test_2, y_train_2, y_test_2),
            'classifier3': (X_train_3, X_test_3, y_train_3_mapped, y_test_3_mapped),
            'full_test': (X_test, y_test),  # For final hierarchical evaluation
            'disease_label_map': disease_label_map
        }
        
        return datasets

    def train_single_classifier(self, X_train, y_train, X_test, y_test, classifier_name, class_labels=None):
        """Train and evaluate a single classifier"""
        print(f"\n{'='*20} {classifier_name} {'='*20}")
        print(f"Training samples: {len(X_train)}")
        print(f"Test samples: {len(X_test)}")
        print(f"Feature vector size: {X_train.shape[1]}")
        
        if class_labels:
            print("Class distribution in training:")
            unique_classes, counts = np.unique(y_train, return_counts=True)
            for cls, count in zip(unique_classes, counts):
                label = class_labels[cls] if cls < len(class_labels) else f"Class_{cls}"
                print(f"  {label}: {count} samples")
        
        # Scale features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Try multiple models
        models = {
            'RandomForest': RandomForestClassifier(n_estimators=100, random_state=SEED, class_weight='balanced'),
            'SVM': SVC(kernel='rbf', random_state=SEED, class_weight='balanced', probability=True),
            'LogisticRegression': LogisticRegression(random_state=SEED, class_weight='balanced', max_iter=1000)
        }
        
        print(f"\n🔍 CROSS-VALIDATION RESULTS:")
        best_score = 0
        best_model = None
        best_model_name = ""
        
        scoring = 'f1_weighted' if len(np.unique(y_train)) > 2 else 'f1'
        
        for name, model in models.items():
            try:
                scores = cross_val_score(model, X_train_scaled, y_train, cv=3, scoring=scoring)
                print(f"  {name:15} | CV F1: {scores.mean():.4f} ± {scores.std():.4f}")
                if scores.mean() > best_score:
                    best_score = scores.mean()
                    best_model = model
                    best_model_name = name
            except Exception as e:
                print(f"  {name:15} | Error: {str(e)}")
        
        if best_model is None:
            print("❌ No model could be trained successfully!")
            return None, None
        
        print(f"\n🏆 BEST MODEL: {best_model_name} (CV F1: {best_score:.4f})")
        
        # Train best model
        best_model.fit(X_train_scaled, y_train)
        
        # Test set evaluation
        y_pred = best_model.predict(X_test_scaled)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        
        if len(np.unique(y_train)) == 2:  # Binary classification
            precision = precision_score(y_test, y_pred)
            recall = recall_score(y_test, y_pred)
            f1 = f1_score(y_test, y_pred)
            print(f"\n📊 TEST PERFORMANCE:")
            print(f"  Accuracy:  {accuracy:.4f}")
            print(f"  Precision: {precision:.4f}")
            print(f"  Recall:    {recall:.4f}")
            print(f"  F1-Score:  {f1:.4f}")
        else:  # Multi-class classification
            f1_macro = f1_score(y_test, y_pred, average='macro')
            f1_weighted = f1_score(y_test, y_pred, average='weighted')
            print(f"\n📊 TEST PERFORMANCE:")
            print(f"  Accuracy:     {accuracy:.4f}")
            print(f"  F1-Macro:     {f1_macro:.4f}")
            print(f"  F1-Weighted:  {f1_weighted:.4f}")
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        print(f"\n📋 CONFUSION MATRIX:")
        
        if class_labels and len(class_labels) == len(np.unique(np.concatenate([y_train, y_test]))):
            display_labels = class_labels
        else:
            display_labels = [f"Class_{i}" for i in range(len(np.unique(np.concatenate([y_train, y_test]))))]
        
        print("Predicted →")
        print("Actual ↓")
        
        # Print confusion matrix with labels
        max_label_len = max(len(str(label)) for label in display_labels)
        header = " " * (max_label_len + 2) + "".join(f"{label:>8}" for label in display_labels)
        print(header)
        
        for i, (actual_label, row) in enumerate(zip(display_labels, cm)):
            row_str = f"{actual_label:>{max_label_len}}: " + "".join(f"{val:>8}" for val in row)
            print(row_str)
        
        # Classification Report
        print(f"\n📈 DETAILED CLASSIFICATION REPORT:")
        print(classification_report(y_test, y_pred, target_names=display_labels, digits=4))
        
        return best_model, scaler

    def train_classifiers(self, datasets):
        """Train all three classifiers"""
        print("\n" + "="*70)
        print("TRAINING HIERARCHICAL CLASSICAL ML CLASSIFIERS")
        print("="*70)
        
        # Classifier 1: Crop vs Weed
        X_train_1, X_test_1, y_train_1, y_test_1 = datasets['classifier1']
        self.classifier1, self.scaler1 = self.train_single_classifier(
            X_train_1, y_train_1, X_test_1, y_test_1,
            "CLASSIFIER 1: CROP vs WEED",
            class_labels=["Weed", "Crop"]
        )
        
        # Classifier 2: Healthy vs Disease
        X_train_2, X_test_2, y_train_2, y_test_2 = datasets['classifier2']
        if len(X_train_2) > 0 and len(np.unique(y_train_2)) > 1:
            self.classifier2, self.scaler2 = self.train_single_classifier(
                X_train_2, y_train_2, X_test_2, y_test_2,
                "CLASSIFIER 2: HEALTHY vs DISEASE",
                class_labels=["Disease", "Healthy"]
            )
        else:
            print("\n❌ CLASSIFIER 2: Not enough samples for Healthy vs Disease classification")
            self.classifier2, self.scaler2 = None, None
        
        # Classifier 3: Disease Types
        X_train_3, X_test_3, y_train_3, y_test_3 = datasets['classifier3']
        if len(X_train_3) > 0 and len(np.unique(y_train_3)) > 1:
            disease_class_names = [self.class_names[self.disease_indices[i]] for i in range(len(self.disease_indices))]
            self.classifier3, self.scaler3 = self.train_single_classifier(
                X_train_3, y_train_3, X_test_3, y_test_3,
                "CLASSIFIER 3: DISEASE CLASSIFICATION",
                class_labels=disease_class_names
            )
        else:
            print("\n❌ CLASSIFIER 3: Not enough samples for Disease type classification")
            self.classifier3, self.scaler3 = None, None

    def predict_hierarchical(self, features):
        """Make hierarchical predictions"""
        predictions = []
        prediction_paths = []  # Track decision path for each sample
        
        for i in range(len(features)):
            sample = features[i:i+1]
            path = []
            
            # Step 1: Crop vs Weed
            sample_scaled_1 = self.scaler1.transform(sample)
            is_crop = self.classifier1.predict(sample_scaled_1)[0]
            crop_prob = self.classifier1.predict_proba(sample_scaled_1)[0]
            
            if is_crop == 0:  # Weed
                predictions.append(self.weed_idx)
                path.append(f"Step1: Weed (prob={crop_prob[0]:.3f})")
            else:  # Crop
                path.append(f"Step1: Crop (prob={crop_prob[1]:.3f})")
                
                # Step 2: Healthy vs Disease
                if self.classifier2 is not None:
                    sample_scaled_2 = self.scaler2.transform(sample)
                    is_healthy = self.classifier2.predict(sample_scaled_2)[0]
                    healthy_prob = self.classifier2.predict_proba(sample_scaled_2)[0]
                    
                    if is_healthy == 1:  # Healthy
                        predictions.append(self.healthy_idx)
                        path.append(f"Step2: Healthy (prob={healthy_prob[1]:.3f})")
                    else:  # Disease
                        path.append(f"Step2: Disease (prob={healthy_prob[0]:.3f})")
                        
                        # Step 3: Disease type
                        if self.classifier3 is not None:
                            sample_scaled_3 = self.scaler3.transform(sample)
                            disease_type = self.classifier3.predict(sample_scaled_3)[0]
                            disease_probs = self.classifier3.predict_proba(sample_scaled_3)[0]
                            max_prob = np.max(disease_probs)
                            
                            predictions.append(self.disease_indices[disease_type])
                            disease_name = self.class_names[self.disease_indices[disease_type]]
                            path.append(f"Step3: {disease_name} (prob={max_prob:.3f})")
                        else:
                            # Fallback: predict most common disease
                            predictions.append(self.disease_indices[0])
                            path.append("Step3: Fallback to first disease class")
                else:
                    # Fallback if classifier2 not available
                    predictions.append(self.healthy_idx)
                    path.append("Step2: Fallback to healthy")
            
            prediction_paths.append(" → ".join(path))
        
        return np.array(predictions), prediction_paths

    def evaluate_hierarchical(self, datasets):
        """Evaluate the hierarchical system"""
        print("\n" + "="*70)
        print("HIERARCHICAL SYSTEM EVALUATION")
        print("="*70)
        
        X_test, y_test = datasets['full_test']
        print(f"📊 Evaluating on {len(X_test)} test samples")
        print(f"Test set class distribution:")
        for i, class_name in enumerate(self.class_names):
            count = np.sum(y_test == i)
            percentage = (count / len(y_test)) * 100
            print(f"  - {class_name}: {count} samples ({percentage:.1f}%)")
        
        # Get predictions with timing
        start_time = time.time()
        y_pred, prediction_paths = self.predict_hierarchical(X_test)
        prediction_time = time.time() - start_time
        
        print(f"\n⏱️  Prediction time: {prediction_time:.2f} seconds ({prediction_time/len(X_test)*1000:.2f} ms per image)")
        
        # Overall accuracy
        overall_accuracy = accuracy_score(y_test, y_pred)
        print(f"\n🎯 OVERALL HIERARCHICAL ACCURACY: {overall_accuracy:.4f} ({overall_accuracy*100:.2f}%)")
        
        # Hierarchical decision flow analysis
        print(f"\n🔍 HIERARCHICAL DECISION FLOW ANALYSIS:")
        stage1_crop = 0
        stage1_weed = 0
        stage2_healthy = 0
        stage2_disease = 0
        stage3_predictions = []
        
        for i in range(len(X_test)):
            sample = X_test[i:i+1]
            
            # Stage 1: Crop vs Weed
            sample_scaled_1 = self.scaler1.transform(sample)
            is_crop = self.classifier1.predict(sample_scaled_1)[0]
            
            if is_crop == 0:  # Weed
                stage1_weed += 1
            else:  # Crop
                stage1_crop += 1
                
                # Stage 2: Healthy vs Disease
                if self.classifier2 is not None:
                    sample_scaled_2 = self.scaler2.transform(sample)
                    is_healthy = self.classifier2.predict(sample_scaled_2)[0]
                    
                    if is_healthy == 1:  # Healthy
                        stage2_healthy += 1
                    else:  # Disease
                        stage2_disease += 1
                        
                        # Stage 3: Disease type
                        if self.classifier3 is not None:
                            sample_scaled_3 = self.scaler3.transform(sample)
                            disease_type = self.classifier3.predict(sample_scaled_3)[0]
                            disease_probs = self.classifier3.predict_proba(sample_scaled_3)[0]
                            stage3_predictions.append((disease_type, np.max(disease_probs)))
        
        print(f"  Stage 1 - Classified as CROP: {stage1_crop}/{len(X_test)} ({stage1_crop/len(X_test)*100:.1f}%)")
        print(f"  Stage 1 - Classified as WEED: {stage1_weed}/{len(X_test)} ({stage1_weed/len(X_test)*100:.1f}%)")
        
        if stage1_crop > 0:
            print(f"  Stage 2 - Classified as HEALTHY: {stage2_healthy}/{stage1_crop} ({stage2_healthy/stage1_crop*100:.1f}% of crops)")
            print(f"  Stage 2 - Classified as DISEASED: {stage2_disease}/{stage1_crop} ({stage2_disease/stage1_crop*100:.1f}% of crops)")
        
        if stage3_predictions:
            avg_disease_confidence = np.mean([conf for _, conf in stage3_predictions])
            print(f"  Stage 3 - Average disease classification confidence: {avg_disease_confidence:.3f}")
            
            # Disease type distribution
            disease_type_counts = Counter([dtype for dtype, _ in stage3_predictions])
            print(f"  Stage 3 - Disease type predictions:")
            for disease_type, count in disease_type_counts.most_common():
                disease_name = self.class_names[self.disease_indices[disease_type]]
                print(f"    {disease_name}: {count} predictions")
        
        # Per-class performance analysis
        print(f"\n📈 PER-CLASS PERFORMANCE ANALYSIS:")
        for i, class_name in enumerate(self.class_names):
            if np.sum(y_test == i) > 0:  # Only analyze classes present in test set
                true_positives = np.sum((y_test == i) & (y_pred == i))
                false_positives = np.sum((y_test != i) & (y_pred == i))
                false_negatives = np.sum((y_test == i) & (y_pred != i))
                
                precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
                recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
                f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
                
                support = np.sum(y_test == i)
                print(f"  {class_name:25} | Precision: {precision:.3f} | Recall: {recall:.3f} | F1: {f1:.3f} | Support: {support}")
        
        # Classification report
        print(f"\n📊 DETAILED CLASSIFICATION REPORT:")
        print("-" * 70)
        print(classification_report(y_test, y_pred, target_names=self.class_names, digits=4))
        
        # Error analysis
        print(f"\n❌ ERROR ANALYSIS:")
        errors = y_test != y_pred
        error_count = np.sum(errors)
        print(f"Total errors: {error_count}/{len(y_test)} ({error_count/len(y_test)*100:.2f}%)")
        
        if error_count > 0:
            print("Most common misclassifications:")
            error_pairs = list(zip(y_test[errors], y_pred[errors]))
            error_counter = Counter(error_pairs)
            for (true_class, pred_class), count in error_counter.most_common(5):
                true_name = self.class_names[true_class]
                pred_name = self.class_names[pred_class]
                print(f"  {true_name} → {pred_name}: {count} times ({count/error_count*100:.1f}% of errors)")
        
        # Confusion Matrix Visualization
        print(f"\n📋 OVERALL CONFUSION MATRIX:")
        cm = confusion_matrix(y_test, y_pred, labels=list(range(len(self.class_names))))
        
        # Print text version
        max_name_len = max(len(name) for name in self.class_names)
        header = " " * (max_name_len + 2) + "".join(f"{i:>6}" for i in range(len(self.class_names)))
        print("Predicted →")
        print("Actual ↓")
        print(header)
        
        for i, (class_name, row) in enumerate(zip(self.class_names, cm)):
            row_str = f"{class_name:>{max_name_len}}: " + "".join(f"{val:>6}" for val in row)
            print(row_str)
        
        # Plot confusion matrix
        plt.figure(figsize=(12, 10))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=self.class_names)
        disp.plot(xticks_rotation=45)
        plt.title('Hierarchical Classical ML - Overall Confusion Matrix', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        # Performance summary
        macro_f1 = f1_score(y_test, y_pred, average='macro')
        weighted_f1 = f1_score(y_test, y_pred, average='weighted')
        
        print(f"\n🏆 FINAL PERFORMANCE SUMMARY:")
        print("="*50)
        print(f"Overall Accuracy:    {overall_accuracy:.4f}")
        print(f"Macro F1-Score:      {macro_f1:.4f}")
        print(f"Weighted F1-Score:   {weighted_f1:.4f}")
        print(f"Prediction Speed:    {prediction_time/len(X_test)*1000:.2f} ms/image")
        print("="*50)
        
        # Show some example predictions with their paths
        print(f"\n🔍 EXAMPLE PREDICTION PATHS (first 10 samples):")
        for i in range(min(10, len(y_test))):
            true_name = self.class_names[y_test[i]]
            pred_name = self.class_names[y_pred[i]]
            correct = "✓" if y_test[i] == y_pred[i] else "✗"
            print(f"  {correct} True: {true_name:25} | Pred: {pred_name:25}")
            print(f"    Path: {prediction_paths[i]}")
        
        return y_test, y_pred

    def save_models(self, save_dir="models/classical_ml/"):
        """Save trained models and scalers"""
        os.makedirs(save_dir, exist_ok=True)
        
        if self.classifier1 is not None:
            joblib.dump(self.classifier1, os.path.join(save_dir, 'classifier1.pkl'))
            joblib.dump(self.scaler1, os.path.join(save_dir, 'scaler1.pkl'))
        
        if self.classifier2 is not None:
            joblib.dump(self.classifier2, os.path.join(save_dir, 'classifier2.pkl'))
            joblib.dump(self.scaler2, os.path.join(save_dir, 'scaler2.pkl'))
        
        if self.classifier3 is not None:
            joblib.dump(self.classifier3, os.path.join(save_dir, 'classifier3.pkl'))
            joblib.dump(self.scaler3, os.path.join(save_dir, 'scaler3.pkl'))
        
        print(f"💾 Models saved to {save_dir}")

# -----------------
# MAIN EXECUTION
# -----------------
def main():
    print("🌿 LETTUCE DISEASE CLASSIFICATION - HIERARCHICAL CLASSICAL ML")
    print("="*70)
    
    # Dataset configuration
    DATA_DIR = "/kaggle/input/lettuce-diseases/Lettuce_disease_datasets"  # Adjust path as needed
    class_names = ["Bacterial", "Downy_mildew_on_lettuce", "Healthy", 
                   "Powdery_mildew_on_lettuce", "Septoria_blight_on_lettuce", 
                   "Shepherd_purse_weeds", "Viral", "Wilt_and_leaf_blight_on_lettuce"]
    
    print(f"📂 Dataset directory: {DATA_DIR}")
    print(f"🏷️  Expected classes: {len(class_names)}")
    for i, class_name in enumerate(class_names):
        print(f"   {i}: {class_name}")
    
    # Initialize hierarchical classifier
    print(f"\n🔧 Initializing hierarchical classifier...")
    hierarchical_ml = HierarchicalClassicalML(DATA_DIR, class_names)
    
    # Load and extract features
    features, labels = hierarchical_ml.load_and_extract_features()
    
    if features is None or labels is None:
        print("❌ Feature extraction failed. Exiting...")
        return None, None, None
    
    # Prepare datasets
    datasets = hierarchical_ml.prepare_datasets(features, labels)
    
    # Train classifiers
    hierarchical_ml.train_classifiers(datasets)
    
    # Evaluate hierarchical system
    y_true, y_pred = hierarchical_ml.evaluate_hierarchical(datasets)
    
    # Save models
    print(f"\n💾 Saving trained models...")
    hierarchical_ml.save_models()
    
    # Additional analysis
    print(f"\n🔍 ADDITIONAL ANALYSIS:")
    
    # Feature importance (if using Random Forest)
    if hasattr(hierarchical_ml.classifier1, 'feature_importances_'):
        print(f"\n🌟 Top 10 Most Important Features (Classifier 1 - Crop vs Weed):")
        feature_importance_1 = hierarchical_ml.classifier1.feature_importances_
        top_features_1 = np.argsort(feature_importance_1)[-10:][::-1]
        for i, feature_idx in enumerate(top_features_1):
            print(f"   {i+1:2d}. Feature {feature_idx:3d}: {feature_importance_1[feature_idx]:.6f}")
    
    if hierarchical_ml.classifier2 is not None and hasattr(hierarchical_ml.classifier2, 'feature_importances_'):
        print(f"\n🌟 Top 10 Most Important Features (Classifier 2 - Healthy vs Disease):")
        feature_importance_2 = hierarchical_ml.classifier2.feature_importances_
        top_features_2 = np.argsort(feature_importance_2)[-10:][::-1]
        for i, feature_idx in enumerate(top_features_2):
            print(f"   {i+1:2d}. Feature {feature_idx:3d}: {feature_importance_2[feature_idx]:.6f}")
    
    if hierarchical_ml.classifier3 is not None and hasattr(hierarchical_ml.classifier3, 'feature_importances_'):
        print(f"\n🌟 Top 10 Most Important Features (Classifier 3 - Disease Types):")
        feature_importance_3 = hierarchical_ml.classifier3.feature_importances_
        top_features_3 = np.argsort(feature_importance_3)[-10:][::-1]
        for i, feature_idx in enumerate(top_features_3):
            print(f"   {i+1:2d}. Feature {feature_idx:3d}: {feature_importance_3[feature_idx]:.6f}")
    
    # Memory usage
    features_memory = sys.getsizeof(features) / (1024*1024)  # MB
    print(f"\n💾 Memory Usage:")
    print(f"   Feature matrix: {features_memory:.2f} MB")
    print(f"   Features per sample: {features.shape[1]}")
    print(f"   Memory per feature vector: {features_memory*1024/len(features):.2f} KB")
    
    # Summary of hierarchical structure performance
    print(f"\n📊 HIERARCHICAL STRUCTURE SUMMARY:")
    print("-" * 50)
    
    # Test each stage individually
    X_test, y_test = datasets['full_test']
    
    # Stage 1 accuracy (Crop vs Weed)
    _, _, y_train_1, y_test_1 = datasets['classifier1']
    y_pred_1 = hierarchical_ml.classifier1.predict(hierarchical_ml.scaler1.transform(X_test))
    stage1_acc = accuracy_score(y_test_1, y_pred_1)
    print(f"Stage 1 (Crop vs Weed) Accuracy: {stage1_acc:.4f}")
    
    # Stage 2 accuracy (Healthy vs Disease) - only on crop samples
    if hierarchical_ml.classifier2 is not None:
        _, _, y_train_2, y_test_2 = datasets['classifier2']
        X_test_2, _, _, _ = datasets['classifier2']
        y_pred_2 = hierarchical_ml.classifier2.predict(hierarchical_ml.scaler2.transform(X_test_2))
        stage2_acc = accuracy_score(y_test_2, y_pred_2)
        print(f"Stage 2 (Healthy vs Disease) Accuracy: {stage2_acc:.4f}")
    
    # Stage 3 accuracy (Disease types) - only on disease samples
    if hierarchical_ml.classifier3 is not None:
        _, _, y_train_3, y_test_3 = datasets['classifier3']
        X_test_3, _, _, _ = datasets['classifier3']
        y_pred_3 = hierarchical_ml.classifier3.predict(hierarchical_ml.scaler3.transform(X_test_3))
        stage3_acc = accuracy_score(y_test_3, y_pred_3)
        print(f"Stage 3 (Disease Classification) Accuracy: {stage3_acc:.4f}")
    
    overall_accuracy = accuracy_score(y_true, y_pred)
    print(f"Overall Hierarchical System Accuracy: {overall_accuracy:.4f}")
    
    print(f"\n" + "="*70)
    print("🎉 HIERARCHICAL CLASSICAL ML PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*70)
    
    return hierarchical_ml, y_true, y_pred

if __name__ == "__main__":
    hierarchical_ml, y_true, y_pred = main()
```
